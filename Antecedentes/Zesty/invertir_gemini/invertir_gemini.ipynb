```python
# PASO 2: CÁLCULO DE MÉTRICAS CUANTITATIVAS
import pandas as pd
import numpy as np

def zesty_step_2_features(df_precios):
    print("Iniciando Paso 2: Procesamiento matemático para 408 fondos...")
    
    # Pre-asignamos un DataFrame para los resultados
    features = pd.DataFrame(index=df_precios.columns)
    
    # 1. Momentum: Retorno Logarítmico a 3 Meses (aprox. 63 días hábiles)
    # Usamos np.log para normalizar la distribución de los retornos
    features['Ret_3M'] = np.log(df_precios.iloc[-1] / df_precios.iloc[-63])
    
    # 2. Riesgo: Volatilidad Anualizada de los últimos 20 días
    returns_diarios = df_precios.pct_change()
    features['Vol_Anual'] = returns_diarios.tail(20).std() * np.sqrt(252)
    
    # 3. Fuerza Técnica: RSI (Relative Strength Index) a 14 días
    delta = df_precios.diff()
    
    # Separamos ganancias y pérdidas
    gain = delta.where(delta > 0, 0.0)
    loss = -delta.where(delta < 0, 0.0)
    
    # Media móvil exponencial (EMA) para RSI (más preciso que la media simple)
    ema_gain = gain.ewm(alpha=1/14, min_periods=14, adjust=False).mean()
    ema_loss = loss.ewm(alpha=1/14, min_periods=14, adjust=False).mean()
    
    rs = ema_gain.iloc[-1] / ema_loss.iloc[-1]
    features['RSI14'] = 100 - (100 / (1 + rs))
    
    # 4. Score de Riesgo-Retorno (Sharpe Ratio Simplificado)
    # Agregamos una pequeña constante (1e-6) para evitar divisiones por cero si la volatilidad es nula
    features['Score_Zesty'] = features['Ret_3M'] / (features['Vol_Anual'] + 1e-6)
    
    # Limpieza: Eliminamos fondos que no tengan suficiente historial (ej. creados hace menos de 3 meses)
    features_clean = features.dropna()
    
    print(f"--- REPORTE DE INGENIERÍA DE DATOS ---")
    print(f"Activos procesados exitosamente: {len(features_clean)}")
    print(f"Métricas calculadas: Ret_3M, Vol_Anual, RSI14, Score_Zesty")
    
    return features_clean

# EJECUCIÓN PASO 2
# Le pasamos el df_precios que generó exitosamente en el Paso 1
df_features = zesty_step_2_features(df_precios)

# Mostramos el Top 5 de activos con mejor Score de Riesgo-Retorno
print("\nTop 5 Fondos por Score Zesty (Mayor Retorno ajustado por Riesgo):")
print(df_features.sort_values(by='Score_Zesty', ascending=False).head())
```

```python
# PASO 3: ALGORITMO DE CLUSTERING K-MEANS Y SELECCIÓN TÁCTICA
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

def zesty_step_3_ml_clustering(df_features):
    print("Iniciando Paso 3: Segmentación de Mercado con Machine Learning...")
    
    # 1. Preparación de los datos: Normalizamos para que variables con distintas escalas 
    # (como RSI de 0-100 y Retornos en decimales) no sesguen el algoritmo.
    scaler = StandardScaler()
    variables_ml = df_features[['Ret_3M', 'Vol_Anual', 'RSI14']]
    datos_escalados = scaler.fit_transform(variables_ml)
    
    # 2. Instanciación del Modelo K-Means
    # Dividiremos el ecosistema Zesty en 4 regímenes de mercado distintos.
    kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
    df_features['Cluster_ID'] = kmeans.fit_predict(datos_escalados)
    
    # 3. Interpretación de Clústeres (Análisis de Centroides)
    centroides = pd.DataFrame(scaler.inverse_transform(kmeans.cluster_centers_), 
                              columns=['Ret_3M_Medio', 'Vol_Anual_Media', 'RSI14_Medio'])
    
    # Mapeo dinámico de etiquetas basado en la volatilidad y retorno
    etiquetas = {}
    for i in range(4):
        vol = centroides.loc[i, 'Vol_Anual_Media']
        ret = centroides.loc[i, 'Ret_3M_Medio']
        
        if vol < 0.10:
            etiquetas[i] = "Liquidez / Renta Fija (Bajo Riesgo)"
        elif ret > centroides['Ret_3M_Medio'].median() and vol > centroides['Vol_Anual_Media'].median():
            etiquetas[i] = "Crecimiento Agresivo / Momentum"
        elif ret < 0:
            etiquetas[i] = "Activos Deprimidos / Sobreventa"
        else:
            etiquetas[i] = "Core / Renta Variable (Mercado General)"
            
    df_features['Perfil_Zesty'] = df_features['Cluster_ID'].map(etiquetas)
    
    print("\n--- RESUMEN DE CLÚSTERES (K-MEANS) ---")
    print(df_features['Perfil_Zesty'].value_counts())
    
    return df_features

# EJECUCIÓN PASO 3
# Le pasamos el df_features que generamos en el Paso 2
df_ml = zesty_step_3_ml_clustering(df_features)

# Generamos la recomendación final para un presupuesto bajo ($5 - $10)
# Buscamos el mejor activo del clúster de "Core / Renta Variable" (Para crecimiento equilibrado)
recomendacion_crecimiento = df_ml[df_ml['Perfil_Zesty'] == "Core / Renta Variable (Mercado General)"].sort_values(by='Score_Zesty', ascending=False)

print("\n--- ESTRATEGIA ÓPTIMA PARA MICRO-CAPITALES ($10 USD) ---")
print("Para maximizar crecimiento asumiendo riesgo de mercado, el modelo sugiere el líder del clúster Core:")
if not recomendacion_crecimiento.empty:
    print(recomendacion_crecimiento[['Ret_3M', 'Vol_Anual', 'Perfil_Zesty']].head(3))
else:
    # Failsafe en caso de que el mercado esté anómalo
    mejor_general = df_ml[df_ml['Vol_Anual'] > 0.10].sort_values(by='Score_Zesty', ascending=False).head(3)
    print("Mercado en estado anómalo. Líderes de riesgo > 10% volatilidad:")
    print(mejor_general[['Ret_3M', 'Vol_Anual', 'Perfil_Zesty']])
```

```python
# PASO 4: SELECCIÓN TÁCTICA PARA MICRO-CAPITALES (ALFA PURO)
def zesty_step_4_micro_portfolio(df_features):
    print("Iniciando Paso 4: Aislamiento de Renta Variable y Momentum...")
    
    # 1. Filtro Estricto: Eliminamos todo activo "seguro" o de liquidez
    # La volatilidad anualizada debe ser mayor al 12% (0.12) para calificar como acción/ETF variable
    renta_variable = df_features[df_features['Vol_Anual'] >= 0.12]
    
    # 2. Búsqueda de Momentum Sostenible
    # Queremos fondos que estén subiendo (Ret_3M positivo) pero que NO estén 
    # en una burbuja especulativa a punto de explotar (RSI menor a 70)
    candidatos = renta_variable[(renta_variable['Ret_3M'] > 0) & (renta_variable['RSI14'] < 70)]
    
    # 3. Rankeo por Retorno Absoluto
    # Para $10 USD, nos interesa el mayor crecimiento porcentual, no la suavidad del viaje
    top_crecimiento = candidatos.sort_values(by='Ret_3M', ascending=False)
    
    print(f"\n--- FILTRO DE RENTA VARIABLE APLICADO ---")
    print(f"Activos analizados: {len(df_features)}")
    print(f"Activos de Renta Variable que cumplen criterios de Momentum: {len(top_crecimiento)}")
    
    print("\n--- TOP 3 FONDOS PARA INVERSIÓN DE $10 USD ---")
    if not top_crecimiento.empty:
        # Mostramos los 3 mejores para que pueda elegir o diversificar en fracciones
        print(top_crecimiento[['Ret_3M', 'Vol_Anual', 'RSI14', 'Score_Zesty']].head(3))
    else:
        print("El mercado está en extremo pánico. No hay activos de riesgo que cumplan el momentum actual.")
        # Failsafe: mostrar lo menos golpeado
        print(renta_variable.sort_values(by='Ret_3M', ascending=False)[['Ret_3M', 'Vol_Anual', 'RSI14']].head(3))
        
    return top_crecimiento.head(3)

# EJECUCIÓN PASO 4
portafolio_final = zesty_step_4_micro_portfolio(df_features)
```

```python
# PASO 5: MODELO DE SALIDA DINÁMICA (ATR) PARA ETFS APALANCADOS
import yfinance as yf
import pandas as pd

def zesty_step_5_exit_strategy(ticker="GUSH", capital=10.0):
    print(f"Calculando parámetros de salida cuantitativa para {ticker}...")
    
    # Descargamos los últimos 3 meses para tener historia suficiente para el ATR
    data = yf.download(ticker, period="3mo", interval="1d", progress=False)
    
    if isinstance(data.columns, pd.MultiIndex):
        data.columns = data.columns.droplevel(1) # Aplanar columnas si es necesario
        
    # 1. Cálculo del True Range (TR)
    data['H-L'] = data['High'] - data['Low']
    data['H-PC'] = abs(data['High'] - data['Close'].shift(1))
    data['L-PC'] = abs(data['Low'] - data['Close'].shift(1))
    data['TR'] = data[['H-L', 'H-PC', 'L-PC']].max(axis=1)
    
    # 2. Average True Range (ATR de 14 días)
    data['ATR'] = data['TR'].rolling(window=14).mean()
    
    # Extraemos los valores de la última sesión
    precio_actual = float(data['Close'].iloc[-1])
    atr_actual = float(data['ATR'].iloc[-1])
    
    # 3. Multiplicadores Cuantitativos (Zesty Risk Model)
    # Take Profit (Captura de Momentum Extremo) = Precio + 3 ATR
    take_profit = precio_actual + (3 * atr_actual)
    
    # Stop Loss (Protección de Capital) = Precio - 1.5 ATR
    stop_loss = precio_actual - (1.5 * atr_actual)
    
    # 4. Proyecciones de Capital
    retorno_pct = ((take_profit - precio_actual) / precio_actual)
    riesgo_pct = ((precio_actual - stop_loss) / precio_actual)
    
    ganancia_usd = capital * retorno_pct
    perdida_usd = capital * riesgo_pct
    
    print(f"\n--- ESTRATEGIA DE SALIDA: {ticker} (2x Apalancado) ---")
    print(f"Precio Actual en Mercado : ${precio_actual:.2f}")
    print(f"Volatilidad Diaria (ATR) : ${atr_actual:.2f}")
    
    print(f"\n[EJECUCIÓN EN BRÓKER]")
    print(f"🟢 ORDEN LIMIT (Take Profit) : Vender a ${take_profit:.2f}")
    print(f"🔴 ORDEN STOP (Stop Loss)    : Vender a ${stop_loss:.2f}")
    
    print(f"\n[PROYECCIÓN PARA SUS ${capital:.2f} USD]")
    print(f"Si toca Take Profit : Retira ${capital + ganancia_usd:.2f} (Ganancia: +${ganancia_usd:.2f})")
    print(f"Si toca Stop Loss   : Retira ${capital - perdida_usd:.2f} (Pérdida máxima: -${perdida_usd:.2f})")
    print(f"Ratio Riesgo/Beneficio: 1 a 2.0 (Matemáticamente asimétrico a su favor)")

# EJECUCIÓN PASO 5

zesty_step_5_exit_strategy("GUSH", capital=10.0)
```
